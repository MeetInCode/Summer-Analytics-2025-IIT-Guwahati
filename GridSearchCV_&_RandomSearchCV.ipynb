```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species',axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=1)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
#model_svm = SVC(C=100,kernel= "rbf" ,gamma= 'auto')
model_svm = SVC(gamma= 'auto')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
#now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},cv = 5,return_train_score = False)  # cv - crossvalidation
```

```python
classifier.fit(X,y) # hence c=1 and gamma = auto has highest mean score for 5 times cv
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],                   # here there are 4 * 2 = 8 combinations possible
},n_iter = 3,cv = 5,return_train_score = False)  # n_iter will randomly pick 3 combinations from given 8 combinations and return the best from 3 comninations
```

```python
classifier_r.fit(X,y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
